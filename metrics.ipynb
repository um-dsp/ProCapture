##### The Following Represents all the metric experiments discussed in the paper 
 ExpD : Average number of Active Nodes <br>
 ExpE : Average Activations Weight<br>
 ExpF : Always Active Nodes <br>
 ExpG : Frequency Distances<br>
 ExpH : Dispersation Index<br>
 ExpI : Entropy Index<br>

 Use Accessor functions to extract the needed activations weather per label ,epr prediction or get all and run the experiment. The next blocks 
 cover different ways the metric could be used

```python
gt = {'D':[],'E':[],'H':[],'I':[], 'G':[] , 'F':[]}
ben = {'D':[],'E':[],'H':[],'I':[], 'G':[], 'F':[]}
FGSM={'D':[],'E':[],'H':[],'I':[], 'G':[], 'F':[]}
PGD={'D':[],'E':[],'H':[],'I':[], 'G':[], 'F':[]}

FGSM_diff=[]
PGD_diff=[]
adv_diff=[]

PRED_RANGE = 10
nb_samples=1000

for i in range(PRED_RANGE):
    
    ## Loading MNIST activations by prediction
    M_begning_sample_act = M_begning_sample.get_label_by_prediction(target_prediction=i,limit=nb_samples)   
    M_FGSM_sample_act = M_FGSM_sample.get_label_by_prediction(target_prediction=i,limit=nb_samples)
    M_PGD_sample_act = M_PGD_sample.get_label_by_prediction(target_prediction=i,limit=nb_samples)
    M_ground_truth_act = M_ground_truth.get_label_by_prediction(target_prediction=i,limit=nb_samples)
    
    print('label = ',i,'\n')
    print("the average similarity between ben and FGSM is {}% ".format(compare(M_begning_sample_act, M_FGSM_sample_act,nb_sample=nb_samples)))
    print("the average similarity between ben and PGD is {}% ".format(compare(M_begning_sample_act, M_PGD_sample_act,nb_sample=nb_samples)))
    print("the average similarity between PGD and FGSM is {}% ".format(compare(M_FGSM_sample_act, M_PGD_sample_act,nb_sample=nb_samples)))
    
    FGSM_diff.append(avg_act_diff(M_begning_sample_act,M_FGSM_sample_act,nb_sample=nb_samples))
    PGD_diff.append(avg_act_diff(M_begning_sample_act,M_PGD_sample_act,nb_sample=nb_samples))
    adv_diff.append(avg_act_diff(M_FGSM_sample_act,M_PGD_sample_act,nb_sample=nb_samples))
    plotDiff(FGSM_diff[i],PGD_diff[i],adv_diff[i],len(FGSM_diff[i]))
    
    gt['D'].append(expD(M_ground_truth_act))
    ben['D'].append(expD(M_begning_sample_act))
    FGSM['D'].append(expD(M_FGSM_sample_act))
    PGD['D'].append(expD(M_PGD_sample_act))
    
    gt['E'].append(expE(M_ground_truth_act))
    ben['E'].append(expE(M_begning_sample_act))
    FGSM['E'].append(expE(M_FGSM_sample_act))
    PGD['E'].append(expE(M_PGD_sample_act))
    
    gt['H'].append(expH(M_ground_truth_act))
    ben['H'].append(expH(M_begning_sample_act))
    FGSM['H'].append(expH(M_FGSM_sample_act))
    PGD['H'].append(expH(M_PGD_sample_act))
    
    gt['I'].append(expI(M_ground_truth_act))
    ben['I'].append(expI(M_begning_sample_act))
    FGSM['I'].append(expI(M_FGSM_sample_act))
    PGD['I'].append(expI(M_PGD_sample_act))
    
    gt['G'].append(list(expG(M_ground_truth_act)))
    ben['G'].append(list(expG(M_begning_sample_act)))
    FGSM['G'].append(list(expG(M_FGSM_sample_act)))
    PGD['G'].append(list(expG(M_PGD_sample_act)))
    
    gt['F'].append(expF(M_ground_truth_act))
    ben['F'].append(expF(M_begning_sample_act))
    FGSM['F'].append(expF(M_FGSM_sample_act))
    PGD['F'].append(expF(M_PGD_sample_act))
```
